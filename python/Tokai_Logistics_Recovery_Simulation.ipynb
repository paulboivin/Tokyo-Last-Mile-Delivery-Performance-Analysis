```python
import pandas as pd
import numpy as np

print("Libraries loaded successfully.")
```

```python
# Night Window Intervention Simulation
# Models the impact of pre-delivery confirmation systems on night window delivery 
# failures based on published Japanese logistics research showing
# 30-40% failure reduction from SMS and app confirmation systems.

df = pd.read_csv("tokai_logistics_deliveries.csv")

# Establishing night window baseline.
night_deliveries = df[df['Delivery_Window'] == 'Night']

night_total = len(night_deliveries)
night_failed = len(
    night_deliveries[
        night_deliveries['Delivery_Status'] == 'Failed'])
night_failure_rate = (night_failed / night_total) * 100
night_redelivery_cost = night_deliveries[
    'Redelivery_Cost_JPY'].sum()

# Calculating attempt-weighted cost for current state.
night_failed_records = night_deliveries[
    night_deliveries['Delivery_Status'] == 'Failed']
night_attempt_weighted = (
    night_failed_records['Delivery_Attempt_Number'] * 800
).sum()

print("=" * 60)
print("NIGHT WINDOW BASELINE — CURRENT STATE")
print("=" * 60)
print(f"Total Night deliveries:           {night_total:>10,}")
print(f"Failed deliveries:                {night_failed:>10,}")
print(f"Current failure rate:             {night_failure_rate:>9.2f}%")
print(f"Flat redelivery cost:             "
      f"¥{night_redelivery_cost:>10,.0f}")
print(f"Attempt-weighted redelivery cost: "
      f"¥{night_attempt_weighted:>10,.0f}")
print()

# Modelling three intervention scenarios (conservative, expected, optimistic).
print("=" * 60)
print("INTERVENTION SCENARIOS — PRE-DELIVERY CONFIRMATION")
print("=" * 60)
print(f"{'Scenario':<25} {'Failures Prevented':>20} "
      f"{'New Failure Rate':>18} "
      f"{'Annual Savings':>20}")
print("-" * 87)

scenarios = [
    ('Conservative (25%)', 0.25),
    ('Expected (30%)',     0.30),
    ('Optimistic (35%)',   0.35)
]

for label, reduction in scenarios:
    failures_prevented = int(night_failed * reduction)
    new_failure_count = night_failed - failures_prevented
    new_failure_rate = (new_failure_count / night_total) * 100
    
    # Flat model savings.
    flat_savings = failures_prevented * 800
    
    # Attempt-weighted savings.
    # Calculating proportional attempt-weighted savings.
    avg_attempts = night_failed_records[
        'Delivery_Attempt_Number'].mean()
    weighted_savings = failures_prevented * 800 * avg_attempts
    
    flat_str = f"¥{flat_savings:,.0f}"
    weighted_str = f"¥{weighted_savings:,.0f}"
    
    print(f"{label:<25} {failures_prevented:>20,} "
          f"{new_failure_rate:>17.2f}% "
          f"{flat_str:>20}")

print()
print("=" * 60)
print("ATTEMPT-WEIGHTED SAVINGS PROJECTIONS")
print("=" * 60)
print(f"Average attempts per Night failure: {avg_attempts:.2f}")
print()
print(f"{'Scenario':<25} {'Attempt-Weighted Savings':>30}")
print("-" * 60)

for label, reduction in scenarios:
    failures_prevented = int(night_failed * reduction)
    weighted_savings = failures_prevented * 800 * avg_attempts
    weighted_str = f"¥{weighted_savings:,.0f}"
    print(f"{label:<25} {weighted_str:>30}")

print()
print("Note: Projections assume successful implementation of")
print("pre-delivery confirmation reduces customer-absence-driven")
print("failures at rates documented in Japanese logistics research")
print("(Yamato Holdings, Sagawa Express, Japan Logistics Institute).")
```
